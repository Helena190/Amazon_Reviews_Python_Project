```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
amazon_df = pd.read_csv('/home/useradd/Documentos/Amazon_Reviews/Amazon_Reviews_Python_Project/data/processed/amazon_df_EDA.csv')
```

```python
#somente reviews positivas
positive = amazon_df[amazon_df['feedback'] == 1]
positive
```

```python
#somente reviews negativas
negative = amazon_df[amazon_df['feedback'] == 0]
negative
```

```python
#transformando todas as sentenças em uma lista
sentences = amazon_df['verified_reviews'].tolist()
len(sentences)
```

```python
#transformando em uma única string 
sentences_as_one_string = ' '.join(str(sentence) if isinstance(sentence, str) else '' for sentence in sentences)
```

```python
#nuvem de palavra do dataset
from wordcloud import WordCloud

plt.figure(figsize = (20,20))
plt.imshow(WordCloud().generate(sentences_as_one_string))
```
