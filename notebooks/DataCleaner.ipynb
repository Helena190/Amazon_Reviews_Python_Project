```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
import nltk
from nltk.corpus import stopwords
import string
```

```python
amazon_df = pd.read_csv('/home/useradd/Documentos/Amazon_Reviews/Amazon_Reviews_Python_Project/data/processed/amazon_df_EDA.csv')
```

```python
def message_cleaning(message):
    # Verifique se a mensagem é uma string
    if isinstance(message, str):
        # Remover pontuação
        Test_punc_removed = [char for char in message if char not in string.punctuation]
        Test_punc_removed_join = ''.join(Test_punc_removed)
        # Remover stopwords
        Test_punc_removed_join_clean = [word for word in Test_punc_removed_join.split() if word.lower() not in stopwords.words('english')]
        return ' '.join(Test_punc_removed_join_clean)
    else:
        return ''  # Ou qualquer outro valor que faça sentido no seu contexto
```

```python
amazon_df_clean = amazon_df['verified_reviews'].apply(message_cleaning)
```

```python
# Aplicar a limpeza e manter outras colunas
amazon_df['cleaned_reviews'] = amazon_df['verified_reviews'].apply(message_cleaning)
```

```python
print(amazon_df.columns)
```

```python
vectorizer = CountVectorizer(analyzer=message_cleaning, dtype=np.uint8)
amazon_df['verified_reviews'] = amazon_df['verified_reviews'].fillna('')
amazon_countvectorizer = vectorizer.fit_transform(amazon_df['verified_reviews'])
```
